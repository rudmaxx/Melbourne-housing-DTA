## Loading the dataset

```python
RAW_URL = "https://raw.githubusercontent.com/rudmaxx/Melbourne-housing-DTA/refs/heads/main/data/melb_data.csv"

df_origin = pd.read_csv(RAW_URL)
df = df_origin.copy()

```

Problem:

- Incorrect column name (Bedroom2)   

```python
print("Missing values:\n")
missing = df.isna().sum()
missing_pct = (df.isna().mean() * 100).round(1)
missing_report = pd.DataFrame({
    'count': missing,
    '%': missing_pct
})
missing_report.sort_values(["count"], ascending=False)

```

```python
df.describe()
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

df.loc[df["Landsize"] == 0, "Landsize"] = np.nan
df.loc[df["BuildingArea"] == 0, "BuildingArea"] = np.nan
df.loc[df["YearBuilt"] < 1899, "YearBuilt"] = np.nan
```

```python
print("Missing values:\n")
missing = df.isna().sum()
missing_pct = (df.isna().mean() * 100).round(1)
missing_report = pd.DataFrame({
    'count': missing,
    '%': missing_pct
})
missing_report[missing_report['count'] >0].sort_values(["count"], ascending=False)
```
